1.1 Basic Feature extraction





```python
# Write your code here
from six.moves import urllib
import pandas as pd
t=[]
data_link="https://raw.githubusercontent.com/unt-iialab/info5731_spring2021/main/class_exercises/01-05-1%20%20Adams%20v%20Tanner.txt"
train=urllib.request.urlopen(data_link)
for line in train:
  d=line.decode("utf-8").replace('\r\n','')
  if d:t.append(d)
df=pd.DataFrame(t,columns=['Raw Text'])
print(df)
```



*   Number of Sentences




```python
df['Number of Sentences'] = df['Raw Text'].apply(lambda x: len(str(x).split(".")))
df
```



*   Number of words




```python
df['Number of words']=df['Raw Text'].apply(lambda x:len(str(x).split(" ")))
df
```



*   Number of characters




```python
df['Number of Characters']=df['Raw Text'].str.len()
df
```



*   Average word length



```python
def avg_word_l(s):
  w=s.split()
  if len(w)!=0:
    return (sum(len(word) for word in w)/len(w))
  else:
    return None
df['Average Word length']=df['Raw Text'].apply(lambda x:avg_word_l(x))
df
```



*   Number of stopwords




```python
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')

s=stopwords.words('english')
df['No of stopwords']=df['Raw Text'].apply(lambda x: len([x for x in x.split() if x in s]))
df
```



*   Number of special characters



```python
def spl_char(sentence):
  spl=0
  for i in range(len(sentence)):
    if not((sentence[i].isalpha()) and (sentence[i].isdigit())):
      spl+=1
  return spl
df['No of special characters']=df['Raw Text'].apply(lambda x:spl_char(x))
df
```



*  Number of numerics



```python
df['No of numerics']=df['Raw Text'].apply(lambda x:len([x for x in x.split() if x.isdigit()]))
df
```



*   Number of uppercase words

```python
df['No of uppercase words']=df['Raw Text'].apply(lambda x:len([x for x in x.split() if x.isupper()]))
df
```

1.2 Basic Text Pre-processing of text data 



*  Lower casing



```python
df['Lower Case'] = df['Raw Text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df
```



*   Punctuation removal


```python
df['Removal of Punctuation'] = df['Lower Case'].str.replace('[^\w\s]','')
df
```



*  Stopwords removal




```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
df['Stopwords Removal'] = df['Removal of Punctuation'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df
```



*   Frequent words removal



```python
freq_words = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[:10]
print(freq_words)
freq_words_list = list(freq_words.index)
df['After common words removal'] = df['Stopwords Removal'].apply(lambda x: " ".join(x for x in x.split() if x not in freq_words_list))
df
```



*   Rare words removal


```python
rare_words = pd.Series(' '.join(df['Stopwords Removal']).split()).value_counts()[-10:]
print(rare_words)
rare_words_list = list(rare_words.index)
df['After Rare Words removal'] = df['After common words removal'].apply(lambda x: " ".join(x for x in x.split() if x not in rare_words_list))
df
```



*   Spelling correction


```python
from textblob import TextBlob
df['Spelling Correction'] = df['After Rare Words removal'].apply(lambda x: str(TextBlob(x).correct()))
df
```



*   Tokenization


```python
import nltk
nltk.download('punkt')
df['Tokenization']=df['Spelling Correction'].apply(lambda x:TextBlob(x).words)
df
```



*   Stemming


```python
from nltk.stem import PorterStemmer
s=PorterStemmer()
df['Stemming']=df['Tokenization'].apply(lambda x:" ".join([s.stem(word) for word in x]))
df
```



*   Lemmatization


```python
from textblob import Word
import nltk
nltk.download('wordnet')
df['Lemmatization']=df['Stemming'].apply(lambda x:" ".join([Word(word).lemmatize() for word in x.split()]))
df
```

1.3 Saving to a csv file

```python
df.to_csv('Data after cleaning',index=False)
```

1.4 Advance Text Processing


*   Calculating text frequencies



```python
tfreq=(df['Lemmatization']).apply(lambda x:pd.value_counts(x.split(" "))).sum(axis=0).reset_index()
tfreq.columns=['words','tfreq']
tfreq
```

```python
import itertools
import collections
from nltk import ngrams
from nltk.tokenize import word_tokenize

w = []
for s in df['Lemmatization']:
  w.append(word_tokenize(s))
a=[x for x in w if x != []]
i = list(itertools.chain.from_iterable(a))
one_grams = ngrams(i, 1)
collections.Counter(one_grams).most_common(10)
```

```python
bi_grams = nltk.bigrams(i)
freq = nltk.FreqDist(bi_grams)
(freq.most_common(10))
```

```python
tri_grams = nltk.trigrams(i)
freq = nltk.FreqDist(tri_grams)
freq.most_common(10)
```
